#### 7. Handling Missing Values

```python
import numpy as np
```

```python
# replace '?' by 'nan' and update dataframe 
raw_data.replace('?', np.nan, inplace=True)
```

```python
# Returns total missing values from each column name.
raw_data.isnull().sum()
```

```python
# Display missing values %
raw_data.isnull().sum() / raw_data.shape[0] * 100
```

```python
# axis = 1 , Drop columns by using column name.
raw_data.drop(['symboling','normalized-losses'], axis = 1, inplace=True)
```

```python
raw_data['price'].dtype
```

```python
raw_data['price'] = raw_data['price'].astype('float64')
```

```python
vehicle_df = raw_data.copy()
```

```python
vehicle_df['price'].median()
```

```python
# fill missing values by mean of 'price'
vehicle_df['price'] = vehicle_df['price'].fillna(vehicle_df['price'].median())
```

```python
# drop missing rows from all columns
vehicle_df.dropna(inplace=True)
```

```python
vehicle_df.isnull().sum()
```

#### 13. Merge

```python
# Series with country names and GDP per capita
df_series1 = pd.Series({'United States': 75708,'Germany': 47000,'Japan': 42444,'India': 2200,'Brazil': 10200
                }, name='GDP_Per_Capita')

# Series with country names and population
df_series2 = pd.Series({'United States': 331000000,'Germany': 83700000,'Japan': 126000000,'India': 1390000000,
                 'Brazil': 213000000}, name = 'Population')
```

```python
df1 = pd.DataFrame(df_series1).reset_index()
```

```python
# rename existing column
df1.columns = ['Country','GDP_Per_Capita']
```

```python
# rename existing column
df2 = pd.DataFrame(df_series2).reset_index()
df2.columns = ['Country','Population']
```

```python
pd.merge(df1, df2, on='Country',how='inner')
```

```python
pd.merge(df1, df2, on='Country',how='outer')
```

```python
pd.merge(df1, df2, on='Country',how='left')
```

```python
pd.merge(df1, df2, on='Country',how='right')
```
